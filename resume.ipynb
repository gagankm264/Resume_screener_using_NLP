```python
import re

def clean_resume(text):
    # Remove URLs (handles http, https, and URLs at end of string)
    cleantext = re.sub(r'http\S+', '', text)
    cleantext=re.sub('@\S+','',cleantext)
    cleantext=re.sub('#\S+\s','',cleantext)
    cleantext=re.sub('\s+',' ',cleantext)
    return cleantext
```

```python
clean_resume("the i a ## gagan my web https://gagan.com")
```

```python
df['Resume']=df['Resume'].apply(lambda x:clean_resume(x))
```

```python
print(df['Resume'][0])
```

words to categorical values

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category']=le.transform(df['Category'])
```

```python
df['Category'].unique()
```

Vecotorization

```python
df['Resume'][0]
```

```python
# TF-IDF Vectorizer

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words='english')
```

```python
tfidf.fit(df['Resume'])
x=tfidf.transform(df['Resume'])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(x,df['Category'],test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
clf=OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```
